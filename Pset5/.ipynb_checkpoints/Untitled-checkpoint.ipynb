```python
degfield_map = {
    1: 'communications and Journalism',
    2: 'Computer Science and Statistics',
    3: 'Education',
    4: 'Engineering',
    5: 'Social Science and Psychology',
    6: 'Business',
    7: 'Natural Sciences',
    8: 'Humanities, arts and History',
    9: 'Medical-related',
    10: 'All Other',
}
df['mfield1'] = df['mfield'].map(degfield_map)
```

```python
sex_map = {
    0: "M",
    1: "F"
}
df['SEX'] = df['female'].map(sex_map)
```

```python
ctab = pd.crosstab(
    index=df['mfield1'],
    columns=df['SEX'],
    values=df['mfield'],
    margins = True,
    margins_name="Total",
    aggfunc=lambda x: len(x)
)
ctab['F'] = (ctab['F'] / ctab.loc['Total','Total']).round(2) # Measures the weighted average of women in each field. 
ctab['M'] = (ctab['M'] / ctab.loc['Total','Total']).round(2)
ctab
```

These numbers are specified in percentages. For example, Females who studied business make up 12% of the sample.

We can see that there are numerous degrees that give rise to quite significant differences in logwages. 

```python
# Crosstab of average wage by gender and degree field
ctab1 = pd.crosstab(
    index=df['mfield1'],
    columns=df['SEX'],
    values=df['logwage'],
    aggfunc=np.mean
)
ctab1['Gap'] = ctab1['M']*ctab['M'] - ctab1['F']*ctab['F']
ctab1.sort_values('Gap', ascending=True).head(n=10)
```

```python
ctab = pd.DataFrame(ctab)
ctab1 = pd.DataFrame(ctab1)
```

```python
ctab1.sort_values('Gap', ascending=False)
```

```python
ctab1['Fcf'] = ctab1['F']*ctab['M']
ctab1['mW'] = ctab1['M']*ctab['M']
ctab1['Gap_Fcf'] = ctab1.mW - ctab1.Fcf
```

What we have shown is that there is on the onset, a quite remarkable difference in mean wages of women and mean wages of men in our sample. Accordingly, we can see that a remarkable portion of this gap is explained by distributional differences within fields between men and women. Let's take a close look at the gap within the field of business. From the gap 0.235740, roughly 0.2121 logpoints explains this gap. Put it differently, roughly 20% of the gender gap within business are due to the fact more men are employed in business than women.

```python
gap = ctab1[['Gap', 'Gap_Fcf']]
gap.sort_values('Gap', ascending = False, inplace = True)
```

```python
ctab1['F'] = ctab1['F']*ctab['F']
ctab1['Mcf'] = ctab1['M']*ctab['F']
ctab1['GapcM'] = ctab1.Mcf - ctab1.F
```

```python
gap['Gap_Mcf'] = ctab1['GapcM']
```
