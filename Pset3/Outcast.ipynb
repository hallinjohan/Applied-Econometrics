```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
```

```python
df = pd.read_csv("ovb.csv")
df["const"] = 1
df.head()
```

```python
dfwomen = df[df.female == 1]
dfmen = df[df.female == 1]
```

## Logwage=constant, immigrantstatus

```python
reg1 = sm.OLS(endog=dfwomen['logwage'], exog=dfwomen[['const', 'imm']],)
results = reg1.fit()
print(results.summary())
```

## logwage=constant,education

```python
reg2 = sm.OLS(endog=dfwomen['logwage'], exog=dfwomen[['const', 'educ']])
results = reg2.fit()
print(results.summary())
```

 ## immigrantstatus=constant,education

```python
reg3 = sm.OLS(endog=dfwomen['imm'], exog=dfwomen[['const', 'educ']])
results = reg3.fit()
print(results.summary())
```

## education=constant,immigrantstatus

```python
reg4 = sm.OLS(endog=dfwomen['educ'], exog=dfwomen[['const', 'imm']])
results = reg4.fit()
print(results.summary())
```

## logwage=constant,education,immigrantstatus

```python
reg5 = sm.OLS(endog=dfwomen['logwage'], exog=dfwomen[['const', 'educ', 'imm']])
results = reg5.fit()
print(results.summary())
```
