```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('rd.csv')
```

Constructing the mean values for covariates for each integer value of PSU

```python
df = df.assign(jump = (df['psu'] - 475)*df['over475'])
buckets = np.arange(300,701,1)
df["Buckets"] = pd.cut(df["psu"], buckets,\
                       labels =np.arange(300,700,1) )

```

```python
mean_values = df.groupby('Buckets').mean()
```

```python
fig, axes = plt.subplots(3, 2, figsize = (25,25))
axes[0,0].plot(mean_values['psu'],mean_values['entercollege'])
axes[1,0].plot(mean_values['psu'],mean_values['hsgpa'])
axes[2,0].plot(mean_values['psu'],mean_values['privatehs'])
axes[0,1].plot(mean_values['psu'],mean_values['hidad'])
axes[1,1].plot(mean_values['psu'],mean_values['himom'])
axes[0,0].set_title('Mean Value entercollege to psu score')
axes[1,0].set_title('Mean Value hsgpa to psu score')
axes[2,0].set_title('Mean Value privatehs to psu score')
axes[0,1].set_title('Mean Value hidad to psu score')
axes[1,1].set_title('Mean Value himom to psu score')
```
