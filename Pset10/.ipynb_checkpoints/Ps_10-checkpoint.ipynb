# Problem Set 10: Prediction

The data set rehosp.csv has 48,470 observations on newly born infants. Thefollowing variables are available:

1) mage = mother's age

2) mom_dropout = 1 if mother has < HS education

3) mom_hs =1 if mother has HS and no college

4) mom_somecoll =1 if mother has 1-3 years of college, no BA or higher

5) mom_college = 1 if mother has BA or higher degree

6) lmean_govins = mean fraction of mothers from same zip code who arecovered by Medicaid (a low income insurance program)

7) bmi = mother's BMI prior to pregnancy (bmi = weight/height-squared)

8) infant's 5 minute apgar score

9) bweight = infant's birthweight in grams

10) hospital = 1 if infant returns to hospital in year after birth

### Feature Selection

```python
state = 42
default_estimators = ['apgar5', 'mage', 'lmean_govins',
                     'mom_dropout', 'mom_hs', 'mom_somecoll',
                      'mom_college', 'bweight', 'bmi']

#data pipeline
import re

def select_cols(df, columns):
    return df.loc[:, columns]

def normalize(df, column):
    if column in df.columns:
        if not (df[column].mean() == 0 and df[column].std() == 1):
            df.loc[:, f'norm_{column}'] = (df[column] - (np.ones(len(df)) * df[column].mean()))/df[column].std()
    return df

def polynomial(df, column,  n):
    df.loc[:, f'{column}^{n}'] = df[column]**n
    return df

def interact(df, column1, column2):
    df.loc[:, f'{column1} * {column2}'] = df[column1]*df[column2]
    return df
    
def inverse_interact(df, column1, column2):
    df.loc[:, f'{column1}/{column2}'] = df[column1]/df[column2]
    return df
    



def pipe_data(df, estimators = default_estimators):
    with pd.option_context('mode.chained_assignment', None):
       # df.dropna(inplace = True)
        df.reset_index(inplace = True)
        
        if 'apgar_under_8' in estimators:
            df['apgar_under_8'] = (df['apgar5']< 8).astype(int)
            
        if 'under_25' in estimators:
            df['under_25'] = (df['mage'] <= 25).astype(int)
        
        if 'over_30' in estimators:
            df['over_30'] = (df['mage'] > 30).astype(int)
            
        if 'any_college' in estimators:
            df['any_college'] = df['mom_somecoll'] + df['mom_college']
            
        if 'bweight_under_2500' in estimators:
            df['bweight_under_2500'] = (df['bweight'] < 2500).astype(int)
         
        if 'bweight_over_4000' in estimators:
            df['bweight_over_4000'] = (df['bweight'] > 4000).astype(int)
        
        #polynomial columns. Taking columns of form <column>^n
        r = re.compile(r"(.*(?=\^))(.*)")
        for match in filter(r.search, estimators):
            df = polynomial(df, r.search(match)[1], r.search(match)[2][1:])
        
        #normalized columns. Taking columns of form norm_<column>
        r = re.compile(r"(?<=^norm_).*")
        for match in filter(r.search, estimators):
            df = normalize(df, r.search(match)[0])
            
        #interaction columns. Taking columns of form <column1> * <column2>
        r = re.compile(r"(\S*?)\s*\*\s*(\S*)")
        for match in filter(r.search, estimators):
            df = interact(df, r.search(match)[1], r.search(match)[2])
        
        r = re.compile(r"(.*?)\/(.*)")
        for match in filter(r.search, estimators):
            df = inverse_interact(df, r.search(match)[1], r.search(match)[2]) 
            
        df.dropna( inplace = True)
    pd.reset_option('mode.chained_assignment')
    
    X = select_cols(df, estimators)
    y = df['hospital']
    return X, y

from sklearn.metrics import mean_squared_error


def test_model(data, model, estimators = None):
    train, val = train_test_split(data)
    
    if not estimators:
        estimators = default_estimators
    
    X_train, y_train = pipe_data(train, estimators)
    X_val, y_val = pipe_data(val, estimators)
    
    model.fit(X = X_train, y = y_train)
    y_hat_train = model.predict(X_train)
    y_hat_val = model.predict(X_val)
    
    t_RMSE = mean_squared_error(y_train, y_hat_train)# ** 0.5
    v_RMSE = mean_squared_error(y_val, y_hat_val)# ** 0.5
    
    print(f'Training Mean Squared Error is: {t_RMSE:.6f}')
    print(f'Validation Mean Squared Error is: {v_RMSE:.6f}')
    print(f'Number of Predicted Hospitalizations(Val): {sum(y_hat_val > cutoff)}')
    

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LassoCV, RidgeCV, LogisticRegressionCV, Lasso, Ridge

cutoff = 0.25

new_estimators = ['apgar_under_8',
                  "under_25", 
                  'norm_bmi',
                  'norm_bweight',
                  'norm_bweight * norm_bmi',
                  'any_college',
                  'any_college * lmean_govins',
                  'bmi * norm_bweight',
                  "norm_lmean_govins",
                  'over_30',
                  'bweight_under_2500', 
                  'apgar_under_8 * bweight_under_2500',
                  "under_25 * lmean_govins", 
                    'norm_bmi',
                    'norm_bweight',
                    'norm_bweight * norm_bmi',
                    'any_college',
                    'any_college * lmean_govins',
                    "mom_dropout * lmean_govins",
                    'bmi * norm_bweight',
                    "norm_lmean_govins",
                    'over_30',
                    "over_30 * lmean_govins",
                    'bweight_under_2500', 
                    'apgar_under_8 * bweight_under_2500',
                    'bweight_over_4000',
                    'apgar_under_8 * bweight_over_4000', 
                    'apgar_under_8 * lmean_govins'
]

updated_estimators = default_estimators + new_estimators

for model, m_name in zip([LogisticRegression(solver = 'liblinear'), LinearRegression(), Lasso(), Ridge(), RandomForestClassifier(n_estimators = 25)], ['Logistic', 'Linear', 'Lasso', 'Ridge', 'Random Forest']):
    print(f"{m_name}: \n")
    for estimators, e_name in zip([default_estimators, updated_estimators], ["Default", "Updated"]):
        print(f" {e_name}: ")
        test_model(original_train_data, model, estimators)
        print("\n")
```

```python
def prep_data(data, estimators = default_estimators, random_state = 42):
    train , val = train_test_split(original_train_data, random_state=random_state)
    
    X_train, y_train = pipe_data(train, estimators)
    X_val, y_val = pipe_data(val, estimators)
    
    return X_train, y_train, X_val, y_val

from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve


def fit_model(model, data, cutoff = 0.5, estimators = default_estimators):
    """
    model - the model to estimate the data on
    
    data - full data set to be split 
    for training and validation
    """
    
    X_train, y_train, X_val, y_val = prep_data(data, estimators)
    
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)#(cutoff < model.predict(X_train)).astype(int)
    y_val_pred = model.predict(X_val)#(cutoff < model.predict(X_val)).astype(int)
    
    return y_train, y_train_pred, y_val, y_val_pred

def create_confusion_matrix(y_train, y_train_pred, y_val, y_val_pred, cutoff = 0.5, normalize = True):
    y_hat_train = (cutoff < y_train_pred).astype(int)
    y_hat_val = (cutoff < y_val_pred).astype(int)
    
        
    cm_t = confusion_matrix(y_train, y_hat_train)
    cm_v = confusion_matrix(y_val, y_hat_val)
    
    if normalize:
        cm_t = cm_t.astype('float') / cm_t.sum(axis=1)[:, np.newaxis]
        cm_v = cm_v.astype('float') / cm_v.sum(axis=1)[:, np.newaxis]
        fmt = ".2f"
        vmin, vmax = 0,1
        
    else:
        fmt = "d"
        vmin, vmax = None, None
        
    fig, axes = plt.subplots(2,1, figsize = (4, 8))
    
    sns.heatmap(cm_t, cmap = sns.light_palette("blue", as_cmap=True),
                vmin = vmin, vmax = vmax, annot = True, fmt = fmt, ax = axes[0])
    axes[0].set_title("TRAINING CONFUSION MATRIX")
    axes[0].set_xlabel("Predicted Value")
    axes[0].set_ylabel("True Value")
    
    sns.heatmap(cm_v, cmap = sns.light_palette("blue", as_cmap=True),
                vmin = vmin, vmax = vmax,annot = True, fmt = fmt, ax = axes[1])
    axes[1].set_title("VALIDATION CONFUSION MATRIX")
    axes[1].set_xlabel("Predicted Value")
    axes[1].set_ylabel("True Value")
    
    plt.subplots_adjust(hspace = 0.4)
    plt.show()
        
def report_statistics(y_train, y_train_pred, y_val, y_val_pred, cutoff = 0.5):
    #Note: report_statistics is checking how well our model is labeling hospital and does not use probabilities
    
    y_train_hat = (y_train_pred > cutoff).astype(int)
    y_val_hat = (y_val_pred > cutoff).astype(int)

    t_RMSE = mean_squared_error(y_train, y_train_hat) #**0.5
    v_RMSE = mean_squared_error(y_val, y_val_hat) #**0.5
    
    t_precision = precision_score(y_train, y_train_hat)
    t_recall = recall_score(y_train, y_train_hat)
    t_accuracy = accuracy_score(y_train, y_train_hat)
    
    v_precision = precision_score(y_val, y_val_hat)
    v_recall = recall_score(y_val, y_val_hat)
    v_accuracy = accuracy_score(y_val, y_val_hat)  
    
    report = pd.DataFrame([[t_RMSE, t_accuracy, t_precision, t_recall],
                           [v_RMSE, v_accuracy, v_precision, v_recall]],
                            columns = ['RMSE', 'Accuracy', 'Precision', 'Recall'], 
                            index = ['Training', 'Validation'])
    display(report)
    
    fpr, tpr, roc_auc = roc_curve(y_val, y_val_pred)
    lw = 2
    plt.plot(fpr, tpr,lw = lw, label = 'ROC curve')
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")
    plt.show()
    
```

### Hyper-Parameter Tuning

Finding the Right Alpha:

```python
X_train, y_train, X_val, y_val = prep_data(original_test_data, updated_estimators)

alphas = np.logspace(-3, -1, 30)
scores = [[], []]

for alpha in alphas:
    mod = Lasso(alpha = alpha)
    mod.fit(X_train, y_train)
    y_val_pred = mod.predict(X_val)
    scores[0].append(mean_squared_error(y_val, y_val_pred))
    
    mod = Ridge(alpha = alpha)
    mod.fit(X_train, y_train)
    y_val_pred = mod.predict(X_val)
    scores[1].append(mean_squared_error(y_val, y_val_pred))
    
sns.lineplot(x = alphas, y = scores[0], color = 'r', label = "Lasso")
sns.lineplot(x = alphas, y = scores[1], color = 'b', label = "Ridge")
plt.legend()
plt.xlabel("Alphas")
plt.ylabel("Mean Squared Error")
plt.show()
```

```python
Lasso_CV = LassoCV(cv = cv)
Ridge_CV = RidgeCV(cv = cv)

Lasso_CV.fit(X_train, y_train)
Ridge_CV.fit(X_train, y_train)

alpha_lasso = Lasso_CV.alpha_
alpha_ridge = Ridge_CV.alpha_

print(f"Best Cross-Validated Lasso Alpha: {Lasso_CV.alpha_}")
print(f"Best Cross-Validated Ridge Alpha: {Ridge_CV.alpha_}")
```

Finding the Best Cutoff:

```python
def adjusted_predict(y_pred, cutoff):
    return (y_pred > cutoff).astype(int)
    
cutoffs = np.arange(0, 1.01, step = 0.01)
scores = [[], []]
for cutoff in cutoffs:
    mod = Lasso(alpha = alpha_lasso)
    mod.fit(X_train, y_train)
    scores[0].append(mean_squared_error(y_val, 
                                        adjusted_predict(mod.predict(X_val), cutoff = cutoff)))
    
    mod = Ridge(alpha = alpha_ridge)
    mod.fit(X_train, y_train)
    scores[1].append(mean_squared_error(y_val, 
                                        adjusted_predict(mod.predict(X_val), cutoff = cutoff)))
    
sns.lineplot(x = cutoffs, y = scores[0], color = 'r', label = "Lasso")
sns.lineplot(x = cutoffs, y = scores[1], color = 'b', label = "Ridge")
plt.legend()
plt.xlabel("Cutoffs")
plt.ylabel("Mean Squared Error")
plt.show()
```

We can see from the graph that choosing a Cutoff at about 0.1 should minimize our MSE while minimizing the tradeoff for flase negatives.

# Resampling:

## Over-sampling
Since our dataset is very imbalanced, it may benefit our model to randomly resample observations from the readmitted class in order to create better predictors. 

```python
resample_estimators = [
                #Default Estimators:
                    'apgar5', 
                    'mage',
                    'lmean_govins',
                    'mom_dropout',
                    'mom_hs',
                    'mom_somecoll',
                    'mom_college',
                    'bweight',
                    'bmi',
                #Updated Estimators:
                    'apgar_under_8', 
                    "under_25", 
                    "under_25 * lmean_govins", 
                    'norm_bmi',
                    'norm_bweight',
                    'norm_bweight * norm_bmi',
                    'any_college',
                    #'any_college * lmean_govins', #<- Bad
                    #"mom_dropout * lmean_govins", #<- Bad
                    #'bmi * norm_bweight', #<- Bad
                    "norm_lmean_govins",
                    'over_30',
                    "over_30 * lmean_govins",
                    'bweight_under_2500', 
                    'apgar_under_8 * bweight_under_2500',
                    'bweight_over_4000',
                    'apgar_under_8 * bweight_over_4000', 
                    #'apgar_under_8 * lmean_govins' <- Bad
]
```

```python
from imblearn.pipeline import make_pipeline
from imblearn.over_sampling import RandomOverSampler
co = 0.5

ols = LinearRegression()
lasso = LassoCV()
ridge = RidgeCV()
logit = LogisticRegression(solver='lbfgs')

X_train, y_train = pipe_data(original_train_data.copy(), estimators = resample_estimators)
X_test, y_test = pipe_data(original_test_data.copy(), estimators = resample_estimators)


for model, m_name in zip([ols, lasso, ridge, logit], 
                         ['OLS', 'Lasso', 'Ridge', 'Logit']):
    
    pipeline = make_pipeline(RandomOverSampler(random_state=state), model)
    pipeline.fit(X_train, y_train)
    y_train_pred, y_test_pred = pipeline.predict(X_train), pipeline.predict(X_test) 
    
    print(f'{m_name}:')
    report_statistics(y_train, y_train_pred, y_test, y_test_pred, cutoff = co)
    create_confusion_matrix(y_train, y_train_pred, y_test, y_test_pred, cutoff = co)
```

## Under-sampling
Since our dataset contains a large number of observations but is very imbalanced, it may be beneficial to our model to under-sample the non-readmitted observations.
